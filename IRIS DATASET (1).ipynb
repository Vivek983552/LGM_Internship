```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns 



import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\vivek kumar\\Iris.dataset_csv.csv")
```

```python
from sklearn .preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Species']= encoder.fit_transform(df['Species'])
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,4))
plt.title('comparison on petal sepal length and width on basis of species')
sns.scatterplot(df['PetalLengthCm'],df['PetalWidthCm'],hue=df['Species'],palette='deep')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.title('comparison on petal sepal length and width on basis of species')
sns.scatterplot(df['SepalLengthCm'],df['SepalWidthCm'],hue=df['Species'],palette='deep')
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,vmin=-1,vmax=1,cmap='viridis')
```

```python
sns.pairplot(df,hue="Species", height=3,palette='deep')
```

```python
y=df['Species']
X=df.iloc[ : , 2:5]
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier 
classifier= RandomForestClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.feature_importances_
```

```python
y_predict=classifier.predict(X_test)
```

```python
y_predict
```

```python
y_actual=np.array(y_test)
y_actual
```

```python
from sklearn.metrics import  classification_report

```

```python
print(classification_report(y_actual,y_predict))
```
